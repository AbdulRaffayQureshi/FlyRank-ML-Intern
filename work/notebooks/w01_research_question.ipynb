* **The Business Decision:** Optimizing the resource allocation budget for a content team's revision schedule.
* **The Operational Action:** An SEO editor reviews a prioritized dashboard queue and assigns a technical content writer to update a specific high-opportunity URL asset.
* **The Cost of a Wrong Recommendation:**
    * *False Positive (Type I Error):* The system flags a stale page for a complete rewrite, but the update yields zero traffic recovery. **Cost:** Wasted copywriting budgets, lost editorial resource hours, and missed opportunity costs.
    * *False Negative (Type II Error):* The model overlooks a declining page that actually possessed massive traffic recovery potential. **Cost:** Permanently bleeding organic ranking visibility to market competitors and losing downstream customer conversions.
* **Why ML is Required:** Search performance degradation is highly non-linear. Simple static filters fail because search traffic trends depend on complex interactions between keyword search volumes, position tiers, and historical click-through rate (CTR) drops. A machine learning framework handles these multi-variable patterns concurrently across thousands of tracks without needing brittle, hand-coded rules.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
import numpy as np

# Load the runnable starter dataset asset using its relative path
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

# Extract backing numbers for the data contract
total_tracks = len(df)
active_pages = df[df["impressions_90d"] > 0]
sv_to_impressions_corr = active_pages["search_volume"].corr(active_pages["impressions_90d"])

print(f"Total Dataset Tracks Available: {total_tracks:,} rows")
print(f"Active Pages Keyword Search Volume vs. Impressions Correlation: {sv_to_impressions_corr:.3f}")
```

* **Total Portfolio Footprint:** The starter playground tracks **{total_tracks}** unique page configurations, establishing a robust statistical sample to analyze search trends.
* **The Core Operational Disconnect:** When filtering for pages actively capturing views (`impressions_90d > 0`), the correlation coefficient between estimated keyword `search_volume` and actual page impressions is remarkably weak (**{sv_to_impressions_corr:.3f}**). 

This low correlation provides mathematical proof that traditional keyword metrics are deceptive standalone signals for content health. Because high keyword volume does not automatically guarantee heavy impression share, a sophisticated multi-variable optimization model is justified over simple rule filters.
